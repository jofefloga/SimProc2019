### Ejercicio 1: 
- Resolver la siguiente ecuación recursiva usando funciones como se vió en clase
$$ D_{n}=(n-1) D_{n-1}+(n-1) D_{n-2} \quad n\ge 3$$
con $D_1=0$ y $D_2 = 1$

```python
# Código de solución
import numpy as np

def f(D1, D2, n):
    D = np.zeros(n)
    D[0] = D1
    D[1] = D2
    
    def f2(i):
        nonlocal D
        D[i]= (i)*D[i-1]+(i)*D[i-2]
    
    [f2(i) for i in range(2,n)]
    return D

Dn=f(0,1,10)
Dn
```

### Ejercicio 4:
- Find all of the words in a string that are less than 4 letters.

```python
# Código de solución 
s = 'Find all of the words in a string that are less than four letters'
s = s.split()
s = list(filter(lambda x: len(x)<4, s))
s
```

### Ejercicio 5:
- Use a dictionary comprehension to count the length of each word in a sentence.

```python
sentence = 'Use a dictionary comprehension to count the length of each word in a sentence'
cont = {}
for i in sentence.split():
    count = cont.get(i, len(i))
    cont[i] = count
cont
```
